## Missing Data
Birth Country and Citizenship were not available for all players

```python
data.loc[:,['representing','birth_country','citizenship']].isnull().agg('sum')
```

```python
(data.loc[:,['representing','birth_country','citizenship']].isnull().agg('sum')/len(data)).round(4)*100
```

```python
complete_data = data.loc[~data.birth_country.isnull(),:]
```

## Born Here, Playing There

```python
complete_data.loc[:,'born_here_playing_there'] = (complete_data.birth_country==complete_data.representing).astype('int8')
complete_data.loc[:,'dual_citizens'] = (complete_data.citizenship==complete_data.representing).astype('int8')
complete_data.loc[complete_data.birth_country=='France',].head(24)
```

```python

(
    (complete_data.birth_country==complete_data.representing).value_counts(normalize=True),
    (complete_data.birth_country==complete_data.representing).value_counts(normalize=False)
)
```
